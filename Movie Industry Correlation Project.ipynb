## Cleaning The Data

```python
# Checking Before and After dropping rows with N/A values
# This was done because there are a lot of misalignment, empty data, and a lot of inconsistencies in the data set.

print("Before:", df.shape)
df = df.dropna()
print("After:", df.shape)
```

```python
# Check if the missing and/or misaligned rows are deleted successfully

df.isnull().sum()
```

```python
# Check data types for our columns

df.dtypes
```

```python
# Change columns data type

df['budget'] = df['budget'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Check if the data types successfully changed
df
```

```python
# create correct year column

df['yearcorrect'] = df['released'].str.extract(r'(\d{4})').astype(int)
```

```python
# check if the year is not the same 

df[df['year'] != df['yearcorrect']]
```

```python
# sort by columns

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

## Finding Correlations In The Data & Making Visualizations

```python
# Questions we need to solve:
# 1. Will Budget have a high correlation or impact to increase gross earnings?
# 2. Company will have a high correlation or impact to increase gross earnings?
```

```python
df
```

```python
# scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film (in hundred million)')
plt.ylabel('Gross Earnings (in billion)')

plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})

```

```python
# calculate correlation between numeric columns from our Dataframe

df.corr(numeric_only=True, method='pearson')
```

```python
# making the heatmap for the correlation

correlation_matrix = df.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Changing "string" column to numerize so we can determine if that impacted to gross earning as well or not
 
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
# Making the heatmap 

correlation_matrix = df_numerized.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
# unstacking dataframe to have better comparison

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sort correlation from lowest to highest

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Show correlation that has the value > 0.5
 
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

```python
# From this we can say that Votes and budget have the highest correlation to gross earnings.

# Answering the previous questions:
# 1. Yes budget have a high correlation or impact in increasing the gross earningns of a movie.
# 2. No, Company doesn't have a correlation on increasing the gross earnings of a movie.
# Added notes: Votes surprisingly have high correlation or impact in effecting a movie's gross earning.
```
